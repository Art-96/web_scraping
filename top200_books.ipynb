```python
import pandas as pd
import csv
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
html=urlopen("https://www.goodreads.com/book/popular_by_date/2019")
soup=BeautifulSoup(html,'html5lib')
```

```python
print(soup.prettify())
```

```python
soup.find('a',{'class':'bookTitle'})
```

```python
# def book_name(soup):
#     return(soup.find('a',{'class':'bookTitle'}).span.get_text())
# book_name(soup)

book_name=[]
booklink = soup.find_all('a', attrs={'class': 'bookTitle'})
for link in booklink:
    book_name.append(link.get_text())
book_name
```

```python
links=[line['href'] for line in soup.table.findAll('a',{'class':'bookTitle'})]
```

```python
bookslink=[]
for i in links:
    bookslink.append('https://www.goodreads.com'+i)
```

```python
bookslink
```

```python
# def book_author(soup):
#     return(soup.find('a',{'class':'authorName'}).span.get_text())
# book_author(soup)

author=[]
book=soup.find_all('a', attrs={'class':'authorName'})
for bd in book:
    author.append(bd.get_text())
    
author
```

```python
linkauthor=[line['href'] for line in soup.table.findAll('a',{'class':'authorName'})]
```

```python
linkauthor
```

```python
df=pd.DataFrame({'Book_Name':book_name,'Author':author,'About_Author':linkauthor,'Book_Lnk':bookslink})
```

```python
print(df)
```

```python
df.to_csv('top_books.csv', sep=',')
```
