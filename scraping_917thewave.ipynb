```python
link_pages=["https://www.917thewave.com.au/directory/category/open-for-business/?pagenum=" + str(i) for i in range(1,10)]
```

```python
link_pages
```

```python
link_product=[]
```

```python
for i in link_pages:
    html=requests.get(i)
    soup=BS(html.text, 'lxml')
    for i in soup.findAll('div', class_="gm-sec-title"):
        for l in i.findAll('a'):
            link_product.append("https://www.917thewave.com.au"+ str(l['href']))
```

```python
link_product
```

```python
def get_name(soup):
    return soup.find('h1').text
```

```python
def get_website(soup):
    if soup.find('p', class_='business-contact').find('a'):
        return soup.find('p', class_='business-contact').a['href']
    else:
        return None
```

```python
def get_phone(soup):
    return re.findall('Phone Number:\s+([0-9]+)',re.sub('\n|\t',' ',soup.find('p', class_='business-contact').text))
```

```python
def get_adress(soup):
    return re.findall('Address:\s+([A-Za-z/0-9\s,]+\s)',re.sub('\n|\t',' ',soup.find('p', class_='business-contact').text))
```

```python
name=[]
website=[]
phone=[]
adress=[]
```

```python
for i in link_product:
    html=requests.get(i)
    soup=BS(html.text, 'lxml')
    name.append(get_name(soup))
    website.append(get_website(soup))
    phone.append(get_phone(soup))
    adress.append(get_adress(soup))
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Name':name, 'Phone_Number':phone, 'Adress':adress, 'Website':website})
```
