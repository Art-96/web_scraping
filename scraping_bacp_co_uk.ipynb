```python
from bs4 import BeautifulSoup as BS
import requests
import builtwith
import re
```

```python
builtwith.parse('https://www.bacp.co.uk/search/Register?skip=0')
```

```python
links=['https://www.bacp.co.uk/search/Register?skip='+str(i) for i in range(0,37840,10)]
```

```python
len(links)
```

```python
html=requests.get('https://www.bacp.co.uk/search/Register?skip=0')
soup=BS(html.text, 'lxml')
```

```python
soup
```

```python
name=[]
membership_number=[]
town=[]
status=[]
```

```python
for li in links[0:30]:
    html=requests.get(li)
    soup=BS(html.text,'lxml')
    for i in soup.findAll('article', class_='search-result--person'):
        name.append(re.sub(r"[\s]", "", i.find('div', class_='search-result__title').text))
        membership_number.append(i.find('div', class_='search-result__type').text)
        town.append(i.findAll('div', class_='search-result__title')[1].text)
        status.append(re.sub(r"[\s]", "", i.find('div', class_='search-result__status').text))
```

```python
df=pd.DataFrame({'Name':name,'Membership_number':membership_number,'Town':town,'Status':status})
```
