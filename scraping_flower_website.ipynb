```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)\
Chrome/80.0.3987.122 Safari/537.36', 'accept': '*/*'}
```

```python
url=('http://g3.no/kulturer/fuchsia-x-hybrida/')
```

```python
html=requests.get(url,headers=headers)
```

```python
soup=BeautifulSoup(html.text,'lxml')
```

```python
print(soup.prettify())
```

```python
links=[line['href'] for line in soup.findAll('a')][13:-4]
```

```python
links
```

```python
url2=('http://g3.no/sorter/bella-miranda/')
```

```python
builtwith.parse('http://g3.no/sorter/bella-miranda/')
```

```python
html2=requests.get(url2,headers=headers)
```

```python
soup2=BeautifulSoup(html2.text,'lxml')
```

```python
print(soup2.prettify())
```

```python
def fl_spec(soup):
    return soup.header.h2.text
```

```python
fl_spec(soup2)
```

```python
def fl_type(soup):
    return soup.header.h1.text
```

```python
fl_type(soup2)
```

```python
def fl_col(soup):
    return re.findall('Farge:\s+([A-Za-z/0-9\s]+)',re.sub('\n|\t',' ',soup.findAll('div',{'class':'medium-6 columns'})[1].p.text))[0]
```

```python
fl_col(soup2)
```

```python
def fl_del_week(soup):
    return re.findall('Leveringsuke:\s+([\d,\d*|a-z0-9-\s]+)Farge:',re.sub('\n|\t',' ',soup.findAll('div',{'class':'medium-6 columns'})[1].p.text))[0].strip()
```

```python
fl_del_week(soup2)
```

```python
def fl_picture(soup):
    return soup.findAll('div',{'class':'medium-6 columns'})[0].img['src'] if soup.findAll('div',{'class':'medium-6 columns'})[0].img else None
```

```python
fl_picture(soup2)
```

```python
def fl_growth(soup):
    return re.findall('Vekstkraft:\s+([a-z0-9\s]+)Antall i brett:',re.sub('\n|\t',' ',soup.findAll('div',{'class':'medium-6 columns'})[1].p.text))[0].strip()
```

```python
fl_growth(soup2)
```

```python
def fl_number(soup):
    return re.findall('Antall i brett:\s+([a-z0-9\s]+)Leveringsuke:',re.sub('\n|\t',' ',soup.findAll('div',{'class':'medium-6 columns'})[1].p.text))[0].strip()
```

```python
fl_number(soup2)
```

```python
Specie_Flower=[]
Type=[]
Colour=[]
Delivery_Week=[]
Link_to_Picture=[]
Quantity_in_width=[]
Growth=[]
Link_to_detail_page=[]
```

```python
for link in links[:10]:
    html=requests.get(link,headers=headers)
    soup=BeautifulSoup(html.text,'lxml')
    Specie_Flower.append(fl_spec(soup))
    Type.append(fl_type(soup))
    Colour.append(fl_col(soup))
    Delivery_Week.append(fl_del_week(soup))
    Link_to_Picture.append(fl_picture(soup))
    Quantity_in_width.append(fl_number(soup))
    Growth.append(fl_growth(soup))
    Link_to_detail_page.append(links)
```

```python
df=pd.DataFrame({'Flower_Specie':Specie_Flower,'Type':Type,'Colour':Colour,'Delivery_Week':Delivery_Week,\
                 'Link_to_Picture':Link_to_Picture,'Quantity_in_width':Quantity_in_width,'Growth':Growth,'Link_to_detail_page':Link_to_detail_page})
```
