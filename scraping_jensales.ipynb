```python
import re 
import requests
from bs4 import BeautifulSoup as BS
```

```python
page_links=['https://www.jensales.com/search?search=manual&']
for i in range(1,799):
    page_links.append('https://www.jensales.com/search?search=manual&page='+str(i))
```

```python
page_links[-1]
```

```python
link=[]
for i in page_links[0:20]:
    html=requests.get(i)
    soup=BS(html.text, 'lxml')
    for ii in soup.findAll('h4'):
        for li in ii.findAll('a'):
            link.append(li['href'])
```

```python
link
```

```python
def get_name(soup):
    return soup.find('h1').text
```

```python
def brand_category(soup):
    return soup.find(string='Brand Category: ').findNext('a').get_text()
```

```python
def product_code(soup):
    return re.sub(r"[Product Code:]", "", str(soup.findAll('ul', class_='list-unstyled')[1].findAll('li')[1].text))
```

```python
def price(soup):
    return soup.find('span', class_='price-old-live').text
```

```python
def img_link(soup):
    return soup.find('a', class_='thumbnail')['href']
```

```python
Price=[]
Name=[]
Image_link=[]
Brand_category=[]
Product_code=[]
```

```python
for ni in link:
    html=requests.get(ni)
    soup=BS(html.text, 'lxml')
    Price.append(price(soup))
    Name.append(get_name(soup))
    Image_link.append(img_link(soup))
    Brand_category.append(brand_category(soup))
    Product_code.append(product_code(soup))
```

```python
bd=pd.DataFrame({'Name':Name,'Price':Price,'Brand_category':Brand_category,'Product_code':Product_code,'Image_link':Image_link,'Link':link})
```
