```python
import requests
```

```python
from bs4 import BeautifulSoup as BS
```

```python
import builtwith
```

```python
builtwith.parse('https://www.propertyfinder.com.lb/en/find-broker/search?page')
```

```python
page_links=['https://www.propertyfinder.com.lb/en/find-broker/search?page']
for i in range(2,10):
    page_links.append('https://www.propertyfinder.com.lb/en/find-broker/search?page='+ str(i))
```

```python
page_links
```

```python
links=[]
for_rent=[]
for_sale=[]
commercial=[]
```

```python
for li in page_links:
    html=requests.get(str(li))
    soup=BS(html.text, 'lxml')
    for i in soup.findAll('a', class_='tiles__tile'):
        for_rent.append(i.findAll('div', class_='counter__count')[0].text)
        for_sale.append(i.findAll('div', class_='counter__count')[1].text)
        commercial.append(i.findAll('div', class_='counter__count')[2].text)
        links.append('https://www.propertyfinder.com.lb'+ str(i['href']))
```

```python
def get_name(soup):
    return soup.find('h1', class_='bio-info__name').text
```

```python
def active(soup):
    return soup.findAll('span', class_='table__column')[1].find('a').text
```

```python
def head(soup):
    return soup.findAll('span', class_='table__column')[3].text
```

```python
def get_telefon(soup):
    return soup.find('span', class_='button__text--is-hidden').text
```

```python
name=[]
active_listins=[]
head_office=[]
telefon_number=[]
```

```python
for i in links:
    html=requests.get(i)
    soup=BS(html.text, 'lxml')
    name.append(get_name(soup))
    active_listins.append(active(soup))
    head_office.append(head(soup))
    telefon_number.append(get_telefon(soup))
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Name':name,'Telefon':telefon_number,'For_rent':for_rent,'Sale':for_sale,'Commercial':commercial,'Active_listins':active_listins,'Head_office':head_office,'Link':links})
```
