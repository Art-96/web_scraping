```python
links=["https://panoramafirm.pl/serwis_komputer%C3%B3w/firmy,"+str(link)+".html" for link in range(1,894)]
```

```python
links
```

```python
links_page=[]
def page_parse(li):
    html=requests.get(li, headers=headers)
    soup=BeautifulSoup(html.text,'lxml')
    for i in soup.findAll('a', class_='business-card-title'):
        links_page.append(i['href'])
    return links_page
    
for i in links:
    page_parse(i)
```

```python
len(links_page)
```

```python
def get_name(soup):
    return soup.find('h1').text
```

```python
def get_adress(soup):
    sou=soup.find('div', class_='col-10').text
    ti=re.sub(r"[\s]", "", str(sou))
    return ti
```

```python
def get_link(soup):
    try:
        return soup.find('div', class_='text-truncate').find('a')['href']
    except:
        return None
```

```python
def get_number(soup):
    sou=soup.find('div',string=re.compile('Telefon')).find_next('div').get_text()
    ti=re.sub(r"[\s]", "", str(sou))
    return ti
```

```python
def get_email(soup):
    try:
        return soup.findAll('div', class_='contact-item')[2].find('a', class_='addax-cs_ip_mod_send_email')['href']
    except:
        return None
```

```python
name=[]
adress=[]
site_links=[]
number=[]
email=[]
```

```python
for i in links_page[0:30]:
    html2=requests.get(i, headers=headers)
    soup2=BeautifulSoup(html2.text,'lxml')
    name.append(get_name(soup2))
    site_links.append(get_link(soup2))
    adress.append(get_adress(soup2))
    number.append(get_number(soup2))
    email.append(get_email(soup2))
```
