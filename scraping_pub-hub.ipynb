```python
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS
```

```python
html=requests.get('https://www.pub-hub.com/index.php/en/shop')
```

```python
soup=BS(html.text, 'lxml')
```

```python
soup
```

```python
page_list=[l['href'] for i in soup.findAll('p', class_='name') for l in i.findAll('a')]
```

```python
page_list
```

```python
html2=requests.get('https://www.pub-hub.com/index.php/en/shop/detail/65')
soup2=BS(html2.text, 'lxml')
```

```python
def get_name(soup):
    return soup.find('h3', class_='shop_name').find('span').text
```

```python
get_name(soup2)
```

```python
def get_adress(soup):
    return soup.find(string='Address').findNext('dd').get_text()
```

```python
get_adress(soup2)
```

```python
def get_station(soup):
    return soup.find(string='Nearest Station').findNext('dd').get_text()
```

```python
get_station(soup2)
```

```python
def get_phone(soup):
    return soup.find(string='Phone').findNext('dd').get_text()
```

```python
get_phone(soup2)
```

```python
def get_capacity(soup):
    return soup.find(string='Seating Capacity').findNext('dd').get_text()
```

```python
get_capacity(soup2)
```

```python
def get_hour(soup):
    if soup.find('div', class_='shop_detail').find(string='HAPPY HOUR'):
        return re.sub(r"\s", "", str(soup.find('div', class_='shop_detail').find(string='HAPPY HOUR').findNext('dd').get_text()))
    else:
        return None
```

```python
get_hour(soup2)
```

```python
def get_time(soup):
    return re.sub(r"\s", "", soup.find(string='Open Time').findNext('dd').get_text())
```

```python
get_time(soup2)
```

```python
name=[]
adress=[]
nearest_station=[]
phone=[]
seating_capacity=[]
open_time=[]
happy_hour=[]
```

```python
for i in page_list:
    html=requests.get(i)
    soup=BS(html.text, 'lxml')
    name.append(get_name(soup))
    adress.append(get_adress(soup))
    nearest_station.append(get_station(soup))
    phone.append(get_phone(soup))
    seating_capacity.append(get_capacity(soup))
    open_time.append(get_time(soup))
    happy_hour.append(get_hour(soup))
```

```python
len(name)
```

```python
len(open_time)
```

```python
df=pd.DataFrame({'Name':name,'Phone':phone,'Adress':adress,'Nearest Station':nearest_station,'Seating_Capacity':seating_capacity,\
                 'Open_Time':open_time,'HAPPY_HOUR':happy_hour})
```
